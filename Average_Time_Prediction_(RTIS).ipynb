**FILTERING DATA**

```python
import pandas as pd

# Take user input for the desired values
desired_event_types = input("Enter desired event types (separated by comma): ").split(",")
desired_vsttn_codes = input("Enter desired VSTTN codes (separated by comma): ").split(",")

# Trim whitespace from the user input
desired_event_types = [event_type.strip() for event_type in desired_event_types]
desired_vsttn_codes = [vsttn_code.strip() for vsttn_code in desired_vsttn_codes]

# Filter the data based on the values in the specified columns
filtered_data = ds[ds['CACEVENTTYPE'].isin(desired_event_types) & ds['CAVSTTNCODE'].isin(desired_vsttn_codes)]

# Print the filtered data
print(filtered_data)
```

```python
da=filtered_data
```

```python
da = pd.DataFrame(da)
input1 = input("Enter the value for : ")
input2 = input("Enter the value for : ")
# Filter rows where CAVSTTNCODE contains both 'CYZ' and 'GZB'
filtered_da = da.groupby('CANDEVICEID').filter(lambda x: input1 in x['CAVSTTNCODE'].values and input2 in x['CAVSTTNCODE'].values)

# Sort the filtered DataFrame based on CANDEVICEID
sorted_da = filtered_da.sort_values(by=['CANDEVICEID','CACEVENTTYPE'])

print(sorted_da)
```

**SORTING DATA**

```python
dt=sorted_da
```

```python
import pandas as pd

dt = pd.DataFrame(dt)
# Filter rows with different 'CAVSTTNCODE' but the same 'CAVLOCONUMB'
filtered_da1 = dt[dt.duplicated(subset=['CAVLOCONUMB'], keep=False) & (dt['CACEVENTTYPE'] != 'A')]

conditions = ((filtered_data['CACEVENTTYPE'] == 'R') & (filtered_data['CACEVENTTYPE'].shift(-1) == 'A')) | \
             ((filtered_data['CACEVENTTYPE'] == 'A') & (filtered_data['CACEVENTTYPE'].shift(-1) == 'R')) | \
             ((filtered_data['CACEVENTTYPE'] == 'R') & (filtered_data['CACEVENTTYPE'].shift(-1) == 'R')) | \
             ((filtered_data['CACEVENTTYPE'] == 'R') & (filtered_data['CACEVENTTYPE'].shift(-1) == 'D')) | \
             ((filtered_data['CACEVENTTYPE'] == 'D') & (filtered_data['CACEVENTTYPE'].shift(-1) == 'R')) | \
             ((filtered_data['CACEVENTTYPE'] == 'A') & (filtered_data['CACEVENTTYPE'].shift(-1) == 'D')) | \
             ((filtered_data['CACEVENTTYPE'] == 'D') & (filtered_data['CACEVENTTYPE'].shift(-1) == 'A'))
# Use loc with the Boolean condition to filter the DataFrame
filtered_result = filtered_da1.loc[conditions].reset_index(drop=True)

# Printing the filtered result
print(filtered_result)
```

```python
dx=filtered_result
```

**Data Splitting**

```python
import pandas as pd
# Convert the data to a DataFrame
dx = pd.DataFrame(dx)

# Group the data by CANDEVICEID, CAVLOCONUMB, and CANMESSAGEID and aggregate the relevant information
grouped_data = dx.groupby(['CANDEVICEID', 'CAVLOCONUMB',]).agg({
    'CACEVENTTYPE': ['first', 'last'],
    'CAVSTTNCODE': ['first', 'last'],
    'CADEVENTPUBTIME': ['first', 'last']
}).reset_index()

# Flatten the MultiIndex columns
grouped_data.columns = ['_'.join(col).strip() for col in grouped_data.columns.values]

# Rename the columns as per the desired output
grouped_data.rename(columns={
    'CAVSTTNCODE_first': 'STARTING PT',
    'CAVSTTNCODE_last': 'ENDING PT',
    'CADEVENTPUBTIME_first': 'CADEVENTPUBTIME_S',
    'CADEVENTPUBTIME_last': 'CADEVENTPUBTIME_E'
}, inplace=True)

# Print the result
print(grouped_data)
```

```python
tc=grouped_data
```

**AVERAGE PREDICTION**

```python
import pandas as pd

# Create a DataFrame
tc = pd.DataFrame(tc)
tc['CADEVENTPUBTIME_S'] = pd.to_datetime(tc['CADEVENTPUBTIME_S'])
tc['CADEVENTPUBTIME_E'] = pd.to_datetime(tc['CADEVENTPUBTIME_E'])
# Calculate the difference between the two columns
tc["TIME_DIFF"] = tc["CADEVENTPUBTIME_E"] - tc["CADEVENTPUBTIME_S"]
# Remove the invalid time difference
tc = tc.loc[tc["TIME_DIFF"].dt.days >= 0]
# Print the DataFrame
print(tc.to_string())
average_time_diff = tc["TIME_DIFF"].mean()

# Print the average
print("AVERAGE TIME:", average_time_diff)
```
